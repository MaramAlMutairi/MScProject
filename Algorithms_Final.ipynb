```python
# a. Separating out Categorical and Numerical Variables
cat_list = []
num_list = []

for variable in dataset.columns:
    if dataset[variable].dtype.name in ['object']:
        cat_list.append(variable)
    else:
        num_list.append(variable)
```

```python
# Correlation

dataset.corr()
```

```python
# 13. Lets check corrleation between Variables
corrmat = dataset.corr()
f, ax = plt.subplots(figsize=(8, 5))
sns.heatmap(corrmat,  annot=True)
plt.show()
```

```python
# Encoding Categorical Features
categorical_features = ['Software Licenses', 'Type of machine', 'GPU', 'File import']
```

```python
# An utility function to create dummy variable
def create_dummies(df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df

# creating dummy variable

for c_feature in categorical_features:
    dataset= create_dummies(dataset, c_feature )
```

```python
dataset.columns
```

### Dividing the dataset into Training and Testing

```python
#Splitting the data

feature_columns = dataset.columns.difference( ['Cost in pounds'])
feature_columns
```

```python
# Spliting data into training and testing - 70% data as training set and 30% as testing set

from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(dataset[feature_columns],
                                                  dataset['Cost in pounds'],
                                                  test_size = 0.3,
                                                  random_state = 123)
```

## Model building

##  Linear Regression 

```python
## Building the model with train set and make predictions on test set

from sklearn.linear_model import LinearRegression
from sklearn import metrics

# regression model
linreg = LinearRegression()
linreg.fit(train_X, train_y)
```

```python
# defining function for mape
def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred - y_true) / y_true)) * 100
```

```python
# Prediction on the Test set
y_pred = linreg.predict(test_X)

# R square score
LR_score = metrics.r2_score(test_y, y_pred)
print(LR_score)

# Root mean square error (RMSE)
LR_rmse = np.sqrt( metrics.mean_squared_error(test_y, y_pred))
print(LR_rmse)

# mean_absolute_error (MAE)
LR_MAE = metrics.mean_absolute_error(test_y, y_pred)
print(LR_MAE)


# mean absolute percentage error (MAPE)
LR_mape = mape(test_y, y_pred)
print(LR_mape)
```

## K Nearest Neighbor (KNN)

###  Error rate for different k values

```python
#import required packages
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt

rmse_val = [] #to store rmse values for different k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(train_X, train_y)  #fit the model
    pred=model.predict(test_X) #make prediction on test set
    error = sqrt(mean_squared_error(test_y, pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , K , 'is:', error)
```

```python
#plotting the rmse values against k values
curve = pd.DataFrame(rmse_val) #elbow curve 
curve.plot()
```

```python
 # KNN model with K = 3
KNN_model = neighbors.KNeighborsRegressor(n_neighbors = 3)

# fitting model on training data
KNN_model.fit(train_X, train_y)
```

###  Prediction on the Test set

```python
#Predicting the test cases
test_y_pred = KNN_model.predict(test_X)

# Value of Test score
KNN_score = KNN_model.score(test_X, test_y)
print(KNN_score)

# Root mean square error (RMSE)
KNN_rmse = np.sqrt( metrics.mean_squared_error(test_y, test_y_pred))
print(KNN_rmse)

# mean_absolute_error (MAE)
KNN_MAE = metrics.mean_absolute_error(test_y, test_y_pred)
print(KNN_MAE)


# mean absolute percentage error (MAPE)
KNN_mape = mape(test_y, test_y_pred)
print(KNN_mape)
```

# Support vector regressor with linear kernel 

```python
from sklearn.svm import SVR
svr_linear = SVR(kernel='linear', gamma='scale', C=1.0, epsilon=0.1)

# Fitting model on training data
svr_linear.fit(train_X, train_y)
```

###  Prediction on the Test set

```python
#Predicting the test cases
svr_test_ypred = svr_linear.predict(test_X)

# Value of Test score
SVM_score = svr_linear.score(test_X, test_y)
print(SVM_score)

# Root mean square error (RMSE)
SVM_rmse = np.sqrt( metrics.mean_squared_error(test_y, svr_test_ypred))
print(SVM_rmse)

# mean_absolute_error (MAE)
SVM_MAE = metrics.mean_absolute_error(test_y, svr_test_ypred)
print(SVM_MAE)

# mean absolute percentage error (MAPE)
SVM_mape = mape(test_y, svr_test_ypred)
print(SVM_mape)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators = 100,
                              criterion = 'mse',
                              random_state = 1,
                              n_jobs = -1)
# Fitting model on training data
forest.fit(train_X, train_y)
```

## Prediction on the Test set

```python
#Predicting the test cases
rf_test_ypred = forest.predict(test_X)

#Value of Test score
RF_score = forest.score(test_X, test_y)
print(RF_score)

# Root mean square error (RMSE)
RF_rmse = np.sqrt( metrics.mean_squared_error(test_y, rf_test_ypred))
print(RF_rmse)

# mean_absolute_error (MAE)
RF_MAE = metrics.mean_absolute_error(test_y, rf_test_ypred)
print(RF_MAE)


# mean absolute percentage error (MAPE)
RF_mape = mape(test_y, rf_test_ypred)
print(RF_mape)
```

```python
# Encoding Categorical Features
categorical_features = ['Software Licenses', 'Type of machine', 'GPU', 'File import']
```

```python
# An utility function to create dummy variable
def create_dummies(df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df

# creating dummy variable

for c_feature in categorical_features:
    dataset10000= create_dummies(dataset10000, c_feature )
```

### Dividing the dataset10000 into Training and Testing

```python
#Splitting the data

feature_columns = dataset10000.columns.difference( ['Cost in pounds'])
feature_columns
```

```python
# Spliting data into training and testing - 70% data as training set and 30% as testing set

from sklearn.model_selection import train_test_split

train_X10, test_X10, train_y10, test_y10 = train_test_split(dataset10000[feature_columns],
                                                  dataset10000['Cost in pounds'],
                                                  test_size = 0.3,
                                                  random_state = 123)
```

## Model building

##  Linear Regression 

```python
## Building the model with train set and make predictions on test set

from sklearn.linear_model import LinearRegression
from sklearn import metrics

# regression model
linreg = LinearRegression()
linreg.fit(train_X10, train_y10)
```

```python
# Prediction on the Test set
y_pred10 = linreg.predict(test_X10)

# R square score
LR_score = metrics.r2_score(test_y10, y_pred10)
print(LR_score)


# Root mean square error (RMSE)
LR_rmse = np.sqrt( metrics.mean_squared_error(test_y10, y_pred10))
print(LR_rmse)

# mean_absolute_error (MAE)
LR_MAE = metrics.mean_absolute_error(test_y10, y_pred10)
print(LR_MAE)

# mean absolute percentage error (MAPE)
LR_mape = mape(test_y10, y_pred10)
print(LR_mape)
```

## K Nearest Neighbor (KNN)

###  Error rate for different k values

```python
#import required packages
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt

rmse_val = [] #to store rmse values for different k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(train_X10, train_y10)  #fit the model
    pred=model.predict(test_X10) #make prediction on test set
    error = sqrt(mean_squared_error(test_y10, pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , K , 'is:', error)
```

```python
#plotting the rmse values against k values
curve = pd.DataFrame(rmse_val) #elbow curve 
curve.plot()
```

```python
 # KNN model with K = 3
KNN_model = neighbors.KNeighborsRegressor(n_neighbors = 3)

# fitting model on training data
KNN_model.fit(train_X10, train_y10)
```

###  Prediction on the Test set

```python
#Predicting the test cases
test_y_pred10 = KNN_model.predict(test_X10)

# Value of Test score
KNN_score = KNN_model.score(test_X10, test_y10)
print(KNN_score)

# Root mean square error (RMSE)
KNN_rmse = np.sqrt( metrics.mean_squared_error(test_y10, test_y_pred10))
print(KNN_rmse)

# mean_absolute_error (MAE)
KNN_MAE = metrics.mean_absolute_error(test_y10, test_y_pred10)
print(KNN_MAE)

# mean absolute percentage error (MAPE)
KNN_mape = mape(test_y10, test_y_pred10)
print(KNN_mape)
```

# Support vector regressor with linear kernel

```python
#Predicting the test cases
svr_test_ypred10 = svr_linear.predict(test_X10)

# Value of Test score
SVM_score = svr_linear.score(test_X10, test_y10)
print(SVM_score)

# Root mean square error (RMSE)
SVM_rmse = np.sqrt( metrics.mean_squared_error(test_y10, svr_test_ypred10))
print(SVM_rmse)

# mean_absolute_error (MAE)
SVM_MAE = metrics.mean_absolute_error(test_y10, svr_test_ypred10)
print(SVM_MAE)

# mean absolute percentage error (MAPE)
SVM_mape = mape(test_y10, svr_test_ypred10)
print(SVM_mape)
```
